# YouTube Scraper From Scratch

## I - Import libraries

```python
import pandas as pd 
from youtube_transcript_api import YouTubeTranscriptApi as yt
```

## II - Import Dataframe

```python
df2 = pd.read_csv('youtube_publications.csv')
df2.head()
```

```python
df2.shape
```

## III - Get transcripts

```python
transcripts = []
counter = 0

for index, row in df2.iterrows():
    try:
        transcript = yt.get_transcripts({row['youtube_id']}, languages = ['fr'])
        transcripts.append(transcript)
        counter += 1
    except:
        print("Could Not Retrieve Transcript")
        transcripts.append(None)
```

```python
print(counter/1254*100) # percentage of transcripts retrieved
```

```python
counter # number of transcripts retrieved
```

```python
df2['transcripts'] = transcripts
df2
```

```python
df2.info()
```

```python
df2 = df2[df2.transcripts.notnull()] # our dataframe without None values
df2
```
